# GLiNER : Generalist and Lightweight model for Named Entity Recognition

GLiNER is a zero-shot Named Entity Recognition (NER) model

Paper: https://arxiv.org/abs/2311.08526 (by Urchade Zaratiana, Nadi Tomeh, Pierre Holat, Thierry Charnois)</br>
Github: https://github.com/urchade/GLiNER

## Importing required Libs

```python
from gliner import GLiNER
```

## Select the Model

**for more models, visit: https://huggingface.co/urchade**

```python
model = GLiNER.from_pretrained("urchade/gliner_largev2")
```

## Input the data

```python
text = """
In the realm of finance, mergers and acquisitions (M&A) play a crucial role in shaping the landscape of businesses worldwide.
Recently, ABC Corporation announced its acquisition of XYZ Enterprises for $1.5 billion, marking a significant milestone in the industry.
The deal, structured primarily as a cash transaction, is anticipated to bolster ABC's market position and expand its product portfolio.
ABC Corporation, a Fortune 500 company headquartered in New York City, specializes in technology solutions for the healthcare sector.
On the other hand, XYZ Enterprises, based in Silicon Valley, is renowned for its innovative software applications catering to the financial services industry.
The acquisition underscores ABC's strategic vision for growth and diversification, consolidating its foothold in key markets."""

labels = ["ORGANIZATION","LOCATION","MONEY","PRODUCT","INDUSTRY"]
```

```python
entities = model.predict_entities(text, labels, threshold=0.4)

for entity in entities:
    print(entity["text"], "=>", entity["label"])
```
